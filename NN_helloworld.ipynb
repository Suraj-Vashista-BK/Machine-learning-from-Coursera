Import tensorflow and numpy


```
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

build the model and check the summary
uses 'stochastic gradient descent' as optimizer
and 'mean squared error'for loss

```
model=tf.keras.Sequential([keras.layers.Dense(1,input_shape=[1])])

model.compile(optimizer='sgd',loss="mean_squared_error")
model.summary()
```

Prepare the input

x and y are related as follows: y=2x-1

```
xs=np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0],dtype=float)
ys=np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0],dtype=float)
```

Train the model for 500 iterations 

```
model.fit(xs,ys,epochs=500)
```

Checking the trained model's results

for 10 as input : y=2(10)-1=19


```
print(model.predict([10.0]))
```
